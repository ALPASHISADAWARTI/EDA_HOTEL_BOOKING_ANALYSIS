```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import datetime as dt
import matplotlib.pyplot as plt
```

```python
# Import the dataset
url = '/content/drive/MyDrive/Colab Notebooks/ALMABETTER/Hotel Bookings.csv'
df = pd.read_csv(url)
```

# *2. Understanding Your Variables- done*

```python
# Datatype of dataset
df.info()
```

```python
# Systematic description of dataset
df.describe()
```

# *3. Data Wrangling- done*

```python
# Find duplicated values
df[df.duplicated()].shape
```

```python
df[df.duplicated()]
```

```python
# Dropping duplicate values
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

```python
# Sorting the null values
df.isnull().sum().sort_values(ascending=False)
```

```python
# Data cleaning (dropping company and agents column)
df.drop(columns=['company','agent'],inplace=True)
```

```python
# Replace the null values under this column with mean value of children.
df['children'].unique()
df['children'].fillna(df['children'].mean(), inplace = True)
```

```python
#replace the missing value with the mode of 'country' column.
df['country'].fillna('others', inplace = True)
```

```python
# Check if all null values are removed
df.isnull().sum().sort_values(ascending = False)[:6]
```

```python
#Combine Total_group_members column for the total numbers of adults,children and babies
df['Total_group_members'] = df.adults + df.children + df.babies
```

```python
#Combine stays_in_weekend_and_weeks column for the total days of stay
df['stays_in_weekend_and_weeks'] = df.stays_in_weekend_nights + df.stays_in_week_nights
```

```python
# Combine day, month and year of arrival
df['arrival_date']=df['arrival_date_day_of_month'].astype(str)+ "" + df['arrival_date_month'].astype(str)+ "" + df['arrival_date_year'].astype(str)
df['arrival_date']=pd.to_datetime(df['arrival_date'])
```

```python
#Adding new column for total no of guests
df['total_guests']= df['adults']+ df['children']+ df['babies']
```

```python
df.head()
```

# *4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables- done*

```python
# Q1 What is the ratio of reservation of City hotel vs resort hotel?
sns.set(rc={'figure.figsize':(14,12)})
graph= sns.countplot(x='hotel',data=df)
graph.set_xlabel('Hotel Type',fontsize=12)
graph.set_ylabel('Reservation Count',fontsize=12)
```

```python
# Q2 Hotel type and cancellations
sns.set(rc={'figure.figsize':(14,12)})
graph= sns.countplot(x='hotel',hue='is_canceled',data=df)
graph.set_xlabel('Hotel Type/Cancellations',fontsize=12)
graph.set_ylabel('Reservation Count',fontsize=12)
```

```python
# Q3 What is the no of adults reserving the hotel ?
sns.set(rc={'figure.figsize':(14,12)})
graph= sns.countplot(x='adults',hue='hotel',data=df)
graph.set_xlabel('number of Adults',fontsize=12)
graph.set_ylabel('Reservation Count',fontsize=12)
```

```python
# Q4 What is the no of families with children reserving the hotel?
sns.set(rc={'figure.figsize':(14,12)})
graph= sns.countplot(x='children',hue='hotel',data=df)
graph.set_xlabel('number of families with children',fontsize=12)
graph.set_ylabel('Reservation Count',fontsize=12)
```

```python
# Q5 Count of Bookings Year-wise
sns.set(rc={'figure.figsize':(14,12)})
graph= sns.countplot(x='arrival_date_year',hue='hotel',data=df)
graph.set_xlabel('arrival_date_year',fontsize=12)
graph.set_ylabel(' Count',fontsize=12)
```

```python
# Q6 Count of Bookings Month-wise
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df['arrival_date_month'] = pd.Categorical(df['arrival_date_month'], categories=months, ordered=True)
sns.set(rc={'figure.figsize':(14,12)})
graph= sns.countplot(x='arrival_date_month',hue='hotel',data=df)
graph.set_xlabel('arrival_date_month',fontsize=12)
graph.set_ylabel(' Count',fontsize=12)
```

```python
# Q7 Count of Bookings day of month wise
sns.set(rc={'figure.figsize':(14,12)})
graph= sns.countplot(x='arrival_date_day_of_month',hue='hotel',data=df,palette='Paired')
graph.set_xlabel('arrival_date_day_of_month',fontsize=12)
graph.set_ylabel(' Count',fontsize=12)


```

```python
# Q8 Country wise distribution of guests.
country_wise_guests = df[df['is_canceled'] == 0]['country'].value_counts().reset_index()
country_wise_guests.columns = ['country', 'No of guests']
country_wise_guests
```

```python
# Q9 How long do people stay ?
df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
df.head()
```

```python
stay = df.groupby(['total_nights', 'hotel']).agg('count').reset_index()
stay = stay.iloc[:, :3]
stay = stay.rename(columns={'is_canceled':'Number of stays'})
stay
```

```python
plt.figure(figsize = (10,5))
sns.barplot(x = 'total_nights', y = 'Number of stays',data= stay,hue='hotel')
```
